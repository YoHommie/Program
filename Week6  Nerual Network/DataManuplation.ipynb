```python
freq_data= pd.read_csv("C:/Users/adity/OneDrive/Desktop/Sixth Semester/CE6018 Seismic Data Analytics/Program/Week6  Nerual Network/transpose.csv")
```

```python

import re

# Convert the 'RSN' column to strings and extract numbers
numbers_only = [int(re.search(r'\d+', str(s)).group()) for s in freq_data["RSN"] if re.search(r'\d+', str(s))]

# Replace 'RSN' column with the extracted numbers
freq_data["RSN"] = numbers_only

## Display the modified DataFrame
# freq_data
```

```python
# Choose the row to sort by (e.g., 'RSN')
row_to_sort_by = 'RSN'

# Sort the DataFrame based on the values in the selected row
sorted_freq = freq_data.sort_values(by=row_to_sort_by)

# Display the sorted DataFrame
sorted_freq
```

```python
flat_data= pd.read_csv("C:/Users/adity/OneDrive/Desktop/Sixth Semester/CE6018 Seismic Data Analytics/Program/Week6  Nerual Network/ngawest.csv",low_memory=False)
flat_data
```

```python
#Getting Speceific Required rows from the Flat file
req_data = flat_data.loc[[i for i in sorted_freq["RSN"]], ["Earthquake Magnitude",
                                                           "Hypocenter Depth (km)", "Joyner-Boore Dist. (km)", "Vs30 (m/s) selected for analysis", "PGA (g)", "PGV (cm/sec)", "T0.010S"]]

#Resetting the index
req_data.reset_index(inplace=True, drop=True)
req_data
```

```python
#Resetting the indexes and merging bothe files to create the required data file
sorted_freq.reset_index(inplace=True, drop=True)
sorted_freq
```

```python
req_resampledFreq=pd.concat([sorted_freq,req_data],axis=1,join="inner")
req_resampledFreq.to_csv("req_resampledTime_Z.csv")
```

```python
import tensorflow as tf
import tensorflow_probability as tfp
import numpy as np

tfd = tfp.distributions
tfpl = tfp.layers

# Generate some synthetic data
np.random.seed(42)
n_samples = 100
X = np.random.randn(n_samples, 4).astype(np.float32)  # 4 input features
y = np.random.randn(n_samples).astype(np.float32)    # 1 output

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(4,)),
    tfpl.DenseVariational(units=10,  # Number of neurons in the hidden layer
                           make_posterior_fn=tfp.layers.util.default_mean_field_normal_fn(), 
                           make_prior_fn=tfp.layers.default_multivariate_normal_fn, 
                           kl_weight=1/n_samples,
                           activation='relu'),
    tfpl.DenseVariational(units=1,  # Number of output neurons
                           make_posterior_fn=tfp.layers.util.default_mean_field_normal_fn(), 
                           make_prior_fn=tfp.layers.default_multivariate_normal_fn,
                           kl_weight=1/n_samples)
])

# Compile the model
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.01), loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=1000, verbose=0)

# Print model summary
model.summary()
```

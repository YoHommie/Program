```python

X_val = ["0", "0.206896552", "0.413793103", "0.620689655", "0.827586207", "1.034482759", "1.24137931", "1.448275862", "1.655172414", "1.862068966",
          "2.068965517", "2.275862069", "2.482758621", "2.689655172", "2.896551724", "3.103448276", "3.310344828", "3.517241379", "3.724137931", "3.931034483"]
Xreq = DataSet[X_val]

selected_columns = ["4.137931034", "4.344827586", "4.551724138", "4.75862069",
                     "4.965517241", "5.172413793", "5.379310345", "5.586206897", "5.793103448", "6"]
Yreq = DataSet[selected_columns]
Yreq
```

```python
corr_mat= Yreq.corr()
corr_mat
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras import models,optimizers
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

import numpy as np


# Normalizing the data
scalerX = MinMaxScaler()
scalerY = MinMaxScaler()
X_normalized = scalerX.fit_transform(Xreq)
Y_normalized = scalerY.fit_transform(Yreq)

# Split the data into training, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_normalized, Y_normalized, test_size=0.2, random_state=0)


# Reshape data for LSTM input (samples, timesteps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1,)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1,)
print(X_train.shape)
# Define LSTM model
model = models.Sequential([
    LSTM(100, input_shape=(X_train.shape[1], 1)),
    Dense(10)
])

```

```python



# Define the MLE loss function
def mle_loss(y_true, y_pred):
    # Calculate the negative log-likelihood
    nll = -tf.reduce_mean(tf.math.log(y_pred))  # For simplicity, assuming y_pred follows a probability distribution
    return nll

# Compile your model with MLE loss function
# model.compile(optimizer='adam', loss=mle_loss)
# Compile model
model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
# model.compile(optimizer='adam', loss=, metrics=['mae'])

# Early stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = model.fit(X_train, y_train, epochs=500, batch_size=32,
                    validation_split=0.25,)
# history = model.fit(X_train, y_train, epochs=500, batch_size=32,
#                     validation_split=0.25, callbacks=[early_stopping])
```

```python

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
# plt.savefig("C:/Users/adity/OneDrive/Desktop/Sixth Semester/CE6018 Seismic Data Analytics/Program/Week6  Nerual Network/figure/")
plt.show()

```

```python
# Plot model parameters
for layer in model.layers:
    weights = layer.get_weights()
    for i in range(len(weights)):
        plt.hist(weights[i].flatten(), bins=30)
        plt.title('Histogram of Layer {} Weights'.format(layer.name))
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        plt.show()
```

```python
# Make predictions
predictions = model.predict(X_test)
predictions_df= pd.DataFrame(np.array(predictions),columns=Yreq.columns)
Fseries=list(predictions_df.columns)
# Plot predictions vs actual values
plt.figure(figsize=(20, 10))
for i in range(len(Fseries)):
    plt.scatter(y_test[:, i], predictions[:, i], label=Fseries[i], alpha=0.5, s=7, marker='o')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predictions vs Actual Values')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

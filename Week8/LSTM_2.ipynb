```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame named 'data'
# Splitting the data into input and output
input_data = DataSet[["0","0.206896552","0.413793103","0.620689655","0.827586207","1.034482759","1.24137931","1.448275862","1.655172414","1.862068966","2.068965517","2.275862069","2.482758621","2.689655172","2.896551724","3.103448276","3.310344828","3.517241379","3.724137931","3.931034483","4.137931034","4.344827586"]].values
output_data = DataSet[["4.551724138","4.75862069","4.965517241","5.172413793","5.379310345","5.586206897","5.793103448","6"]].values

# Reshape the input data into 3D array [samples, timesteps, features]
input_data = input_data.reshape((input_data.shape[0], input_data.shape[1], 1))

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(input_data.shape[1], 1)))
model.add(Dense(8))  # Output layer with 8 neurons
model.compile(optimizer='adam', loss='mse')  # Using mean squared error loss and adam optimizer

# Train the model
history = model.fit(input_data, output_data, epochs=100, batch_size=72, verbose=2)






```

```python
print(output_data.shape)
```

```python
# Plot training loss
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
# Plot model parameters
for layer in model.layers:
    weights = layer.get_weights()
    for i in range(len(weights)):
        plt.hist(weights[i].flatten(), bins=30)
        plt.title('Histogram of Layer {} Weights'.format(layer.name))
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        plt.show()
```

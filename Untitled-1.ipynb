```python
# %%
# X_val=["Earthquake Magnitude","Hypocenter Depth (km)","Joyner-Boore Dist. (km)","Vs30 (m/s) selected for analysis",]
# X_val=["Earthquake Magnitude","Hypocenter Depth (km)","Joyner-Boore Dist. (km)","Vs30 (m/s) selected for analysis",]
X_val = ["0", "0.206896552", "0.413793103", "0.620689655", "0.827586207", "1.034482759", "1.24137931", "1.448275862", "1.655172414", "1.862068966",
          "2.068965517", "2.275862069", "2.482758621", "2.689655172", "2.896551724", "3.103448276", "3.310344828", "3.517241379", "3.724137931", "3.931034483"]
Xreq = DataSet[X_val]
# X_val+=["RSN","T0.010S","PGA (g)","PGV (cm/sec)"]
# selected_columns = DataSet.columns.difference(X_val)

# Create a new DataFrame with the selected columns
# selected_columns = ["4.137931034", "4.344827586", "4.551724138", "4.75862069",
#                      "4.965517241", "5.172413793", "5.379310345", "5.586206897", "5.793103448", "6"]
# Yreq = DataSet[selected_columns]

# Define the columns you want for Yreq
Y_columns = ["4.137931034", "4.344827586", "4.551724138", "4.75862069",
                     "4.965517241", "5.172413793", "5.379310345", "5.586206897", "5.793103448", "6"]
Yreq = DataSet[Y_columns]
Yreq
```

```python
# %%
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras import models,optimizers
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping

import numpy as np


# Normalizing the data
scalerX = MinMaxScaler()
scalerY = MinMaxScaler()
X_normalized = scalerX.fit_transform(Xreq)
Y_normalized = scalerY.fit_transform(Yreq)

# Split the data into training, validation, and test sets
# X_train, X_test, y_train, y_test = train_test_split(
#     X_normalized, Y_normalized, test_size=0.2, random_state=0)


split_ratio = 0.8

# Calculate the index to split the data
split_index = int(len(X_normalized) * split_ratio)

# Split the data
X_train = X_normalized[:split_index]
X_test = X_normalized[split_index:]
y_train = Y_normalized[:split_index]
y_test = Y_normalized[split_index:]


print(X_train.shape[0], 1, X_train.shape[1])

# Reshape data for LSTM input (samples, timesteps, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Define LSTM model
model = models.Sequential([
    LSTM(10, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(30)
])


```

```python
# %%



# Define the MLE loss function
def mle_loss(y_true, y_pred):
    # Calculate the negative log-likelihood
    nll = -tf.reduce_mean(tf.math.log(y_pred))  # For simplicity, assuming y_pred follows a probability distribution
    return nll

# Compile your model with MLE loss function
# model.compile(optimizer='adam', loss=mle_loss)
# Compile model
model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
# model.compile(optimizer='adam', loss=, metrics=['mae'])

# Early stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = model.fit(X_train, y_train, epochs=500, batch_size=32,
                    validation_split=0.25, callbacks=[early_stopping])

# Evaluate model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class CVAE(nn.Module):
    def __init__(self, input_dim, label_dim, latent_dim):
        super(CVAE, self).__init__()
        
        self.input_dim = input_dim
        self.label_dim = label_dim
        self.latent_dim = latent_dim
        
        # Encoder
        self.encoder_input_layer = nn.Linear(input_dim + label_dim, 256)
        self.encoder_hidden_layer = nn.Linear(256, 128)
        self.encoder_mu = nn.Linear(128, latent_dim)
        self.encoder_logvar = nn.Linear(128, latent_dim)
        
        # Decoder
        self.decoder_input_layer = nn.Linear(latent_dim + label_dim, 128)
        self.decoder_hidden_layer = nn.Linear(128, 256)
        self.decoder_output_layer = nn.Linear(256, input_dim)
        
    def encoder(self, x):
        x = F.relu(self.encoder_input_layer(x))
        x = F.relu(self.encoder_hidden_layer(x))
        mu = self.encoder_mu(x)
        logvar = self.encoder_logvar(x)
        return mu, logvar
    
    def decoder(self, z):
        z = F.relu(self.decoder_input_layer(z))
        z = F.relu(self.decoder_hidden_layer(z))
        x_recon = torch.sigmoid(self.decoder_output_layer(z))
        return x_recon
    
    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std
    
    def forward(self, x, y):
        # Concatenate input with labels
        x = torch.cat((x, y), dim=1)
        
        # Encoder
        mu, logvar = self.encoder(x)
        
        # Reparameterization trick
        z = self.reparameterize(mu, logvar)
        
        # Concatenate latent variable with labels
        z = torch.cat((z, y), dim=1)
        
        # Decoder
        x_recon = self.decoder(z)
        
        return x_recon, mu, logvar

# Example usage
input_dim = 52
label_dim = 6
latent_dim = 5

# Initialize the CVAE model
cvae = CVAE(input_dim, label_dim, latent_dim)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(cvae.parameters(), lr=0.001)

# Example data
x = X_train # Input data
y = y_train  # Labels

# Function to plot losses
def plot_losses(losses):
    plt.plot(losses)
    plt.title("Variation of Losses")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.show()

# Function to plot predicted vs actual values for each feature
def plot_predicted_actual(x, x_recon):
    num_features = x.shape[1]
    for i in range(num_features):
        plt.figure(figsize=(8, 6))
        plt.scatter(x[:, i], x_recon[:, i], color='b', label='Predicted vs Actual')
        plt.plot(x[:, i], x[:, i], color='r', label='Actual vs Actual', linestyle='--')
        plt.title(f"Feature {i+1}: Predicted vs Actual")
        plt.xlabel("Actual")
        plt.ylabel("Predicted")
        plt.legend()
        plt.show()

# Training loop
epochs = 1000
losses = []

for epoch in range(epochs):
    optimizer.zero_grad()
    x_recon, mu, logvar = cvae(x, y)
    
    # Calculate reconstruction loss and KL divergence
    recon_loss = criterion(x_recon, x)
    kl_divergence = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    
    # Total loss
    loss = recon_loss + kl_divergence
    losses.append(loss.item())
    
    # Backpropagation
    loss.backward()
    optimizer.step()
    
    # Print progress
    if epoch % 10 == 0:
        print(f"Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}")

# Plot variation of losses
plot_losses(losses)

# Plot predicted vs actual values for each feature
x_recon, _, _ = cvae(x, y)
plot_predicted_actual(x.detach().numpy(), x_recon.detach().numpy())
```
